```python
import numpy as np
import matplotlib.pyplot as plt
border = 1
s0 = 4
s = 5 
sb = 10 #отставание от правой границы
x0 = 2 * s0
y0 = 5 * s0
x1 = 20 * s
y1 = 10 * s
dx = x1 - x0
dy = y1 - y0
k = dy / dx #угол наклона
y_size = min(y0, y1) + dy + sb * border
x_size = min(x0, x1) + dx + sb * border
vp = np.zeros((y_size, x_size))
print(vp.shape) # В случае цветного изображения форма массива (y_size, x_size, 3).
x = x0
while x <= x1:
    y = int(k * x + 2)
    yp = y_size - (y + border)
    vp[yp, x] = 255
    x = x + 1
plt.figure(figsize = (6, 4))
plt.imshow(vp, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
border = 1
s0 = 3
s = 10
sb = 10 #отставание от правой границы
x0 = 2 * s0
y0 = 5 * s0
x1 = 20 * s
y1 = 10 * s
dx = x1 - x0
dy = y1 - y0
k = dy / dx #угол наклона
y_size = min(y0, y1) + dy + sb * border
x_size = min(x0, x1) + dx + sb * border
image = np.zeros((y_size, x_size, 3))
#print(image.shape)
image[y1, x0, 0] = 255
x = x0
step_coror_change = int(255 / dx)
#print(y_size - y0)
iterator = 0
while x <= x1:
    y = int(k * x + 7)
    yp = y_size - (y + border)
    #print(image[yp, x, 1])
    #print(yp, x)
    image[yp, x, 0] = 255 - iterator*step_coror_change
    #print(image[yp, x, 1])
    image[yp, x, 1] = 255 + iterator*step_coror_change
    x = x + 1
    iterator += 1


fig, ax = plt.subplots(1,1,figsize=(20,20))

image = PIL.Image.fromarray(image.astype('uint8'), 'RGB')
ax.imshow(image)
```
